BASIC SINGLETON

```python
class Application:
    _instance = None

    def __new__(cls):
        if cls._instance is None:
            cls._instance = super().__new__(cls)
        return cls._instance
```

```python
class Application:
    def __init__(self):
        self.state = None
```

```python
 
a_app = Application()
b_app = Application()

a_app.state = "running"
b_app.state = "stopped"

print(a_app.state)
print(b_app.state)
```

METACLASS SINGLETON

```python
class Singleton(type):
    _instances = {}

    def __call__(self, *args, **kwds):
        if self not in self._instances:
            self._instances[self] = super().__call__(*args, **kwds)
        return self._instances[self]
```

```python
class Logger():
    def __init__(self):
        self.logs = []
    
    def log(self, message):
        self.logs.append(message)

    def list(self):
        print("Logs: ")
        for each in self.logs:
            print(each)

lg1 = Logger()
lg2 = Logger()

lg1.log("Initializing")
lg2.log("Starting the system")

lg1.list()
print('-' * 20)
lg2.list()
```
